```python
import pandas as pd
import numpy as np

# Load preprocessed dataset
Final_preprocessed_data = pd.read_csv('preprocessed_features.csv')


# taking the average of ages like if the person's age is predicted as twenties(19-29) then he will be told as he is 24  
def age_to_numeric(age_str):
    if '<' in age_str:
        return int(age_str.split(' ')[1]) - 1
    elif 'twenties' in age_str:
        return 24
    elif 'thirties' in age_str:
        return 34
    elif 'fourties' in age_str:
        return 44
    elif 'fifties' in age_str:
        return 54
    elif 'sixties' in age_str:
        return 64
    elif 'seventies' in age_str:
        return 74
    elif 'eighties' in age_str:
        return 84
    elif 'nineties' in age_str:
        return 94
    else:
        return -1  # Handle unknown or invalid values



Final_preprocessed_data['age'] = Final_preprocessed_data['age'].apply(age_to_numeric)

# Separate string columns
string_columns = Final_preprocessed_data[['filename', 'text', 'gender', 'accent']]
numeric_columns = Final_preprocessed_data.drop(['filename', 'text', 'gender', 'accent'], axis=1)

# Define Linear Regression from scratch
class LinearRegression:
    def __init__(self, learning_rate=0.01, n_iterations=1000):
        self.learning_rate = learning_rate
        self.n_iterations = n_iterations
        self.weights = None
        self.bias = None

    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0

        # Gradient Descent
        for _ in range(self.n_iterations):
            y_predicted = np.dot(X, self.weights) + self.bias
            dw = (1 / n_samples) * np.dot(X.T, (y_predicted - y))
            db = (1 / n_samples) * np.sum(y_predicted - y)

            self.weights -= self.learning_rate * dw
            self.bias -= self.learning_rate * db

    def predict(self, X):
        return np.dot(X, self.weights) + self.bias

# Prepare data for training
X = numeric_columns.values
y = Final_preprocessed_data['age'].values

# Initialize and train the model
model = LinearRegression(learning_rate=0.01, n_iterations=1000)
model.fit(X, y)

# Make predictions
predictions = model.predict(X)

# Display sample predictions
print("Sample predictions:", predictions[:5])
```

```python
# Load preprocessed data
data = pd.read_csv("preprocessed_features.csv")
# Map age groups to numerical values for regression
age_mapping = {'teens': 18, 'twenties': 24, 'thirties': 34, 'fourties': 44,
               'fifties': 54, 'sixties': 64, 'seventies': 74, 'eighties': 84,
               'nineties': 90}
data['age'] = data['age'].map(age_mapping)

# Separate features (X) and target variable (y)
X = data.drop(['filename', 'text', 'age', 'gender', 'accent'], axis=1)  # Exclude non-numeric columns
y = data['age']

# Convert to numpy arrays
X = X.to_numpy()
y = y.to_numpy().reshape(-1, 1)  # Reshape y to 2D array

class LinearRegression:
    def __init__(self, learning_rate=0.01, n_iterations=1000):
        self.learning_rate = learning_rate
        self.n_iterations = n_iterations
        self.weights = None
        self.bias = None

    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.weights = np.zeros((n_features, 1))
        self.bias = 0

        for _ in range(self.n_iterations):
            y_predicted = np.dot(X, self.weights) + self.bias
            dw = (1 / n_samples) * np.dot(X.T, (y_predicted - y))
            db = (1 / n_samples) * np.sum(y_predicted - y)

            self.weights -= self.learning_rate * dw
            self.bias -= self.learning_rate * db

    def predict(self, X):
        return np.dot(X, self.weights) + self.bias

```

```python
# Initialize and train the model
model = LinearRegression(learning_rate=0.01, n_iterations=1000)
model.fit(X, y)

# Load testing data
test_data = pd.read_csv("cv-valid-test.csv")

# Map age groups to numerical values for regression (similar to what we did for training data)
test_data['age'] = test_data['age'].map(age_mapping)

# Separate features (X_test) and target variable (y_test)
X_test = test_data.drop(['filename', 'text', 'age', 'gender', 'accent'], axis=1)
y_test = test_data['age']

# Convert to numpy arrays
X_test = X_test.to_numpy()
y_test = y_test.to_numpy().reshape(-1, 1)  # Reshape y_test to 2D array


# Make predictions
predictions = model.predict(X_test)


# Calculate mean squared error (MSE)
mse = np.mean((predictions - y_test) ** 2)

# Calculate R-squared (R^2) coefficient of determination
total_sum_of_squares = np.sum((y_test - np.mean(y_test)) ** 2)
residual_sum_of_squares = np.sum((y_test - predictions) ** 2)
r2 = 1 - (residual_sum_of_squares / total_sum_of_squares)


print("Mean Squared Error:", mse)
print("R² Score:", r2)
print("Coefficients:", model.weights.flatten())
print("Intercept:", model.bias)
```

```python
import pandas as pd
import numpy as np



# Preprocess 'age' column
def preprocess_age(age_str):
    if age_str == 'teens':
        return 18
    elif age_str == 'twenties':
        return 24
    elif age_str == 'thirties':
        return 34
    elif age_str == 'fourties':
        return 44
    elif age_str == 'fifties':
        return 54
    elif age_str == 'sixties':
        return 64
    elif age_str == 'seventies':
        return 74
    elif age_str == 'eighties':
        return 84
    elif age_str == 'nineties':
        return 94
    else:
        return 0  # Handle unknown values if any

# Load the preprocessed data
data = pd.read_csv('preprocessed_features.csv')
    
# Apply preprocessing to 'age' column
data['age'] = data['age'].apply(preprocess_age)

data

# # Separate features (X) and target variable (y) again
# #SETTING X TO ONLY NUMERUIC COLUMNS 
# X = data.drop(columns=['filename', 'text', 'age', 'gender', 'accent'])
# y = data['age']

# Encode categorical variables (gender, accent) using one-hot encoding
data_encoded = pd.get_dummies(data, columns=['gender', 'accent'])

# Define features and target variable
X = data_encoded.drop(['filename', 'text','age'], axis=1)
y = data_encoded['age']
```

```python


# Implement linear regression from scratch
class my_LinearRegression:
    def __init__(self, lr=0.01, n_iters=1000):
        self.lr = lr
        self.n_iters = n_iters
        self.weights = None
        self.bias = None

    def fit(self, X, y):
        # Initialize weights and bias
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0

        # Gradient descent
        for _ in range(self.n_iters):
            y_pred = np.dot(X, self.weights) + self.bias
            dw = (1 / n_samples) * np.dot(X.T, (y_pred - y))
            db = (1 / n_samples) * np.sum(y_pred - y)
            self.weights -= self.lr * dw
            self.bias -= self.lr * db

    def predict(self, X):
        return np.dot(X, self.weights) + self.bias

# Initialize and train the model
model = my_LinearRegression()
model.fit(X, y)

# Optional: Display coefficients and intercept
print("Coefficients:", model.weights)
print("Intercept:", model.bias)
```
